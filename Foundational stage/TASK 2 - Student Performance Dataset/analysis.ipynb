```python
# Task 2 - Student Performance Analysis
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("StudentsPerformance.csv")
df.head()
```

Data Validation

```python
# Check for missing values
df.isnull().sum()
```

```python
# Check for negative marks
(df[['math score','reading score','writing score']] < 0).sum()

```

```python
df = df.dropna()
```

Calculate Average Marks

```python
avg_math = df['math score'].mean()
avg_reading = df['reading score'].mean()
avg_writing = df['writing score'].mean()

avg_math, avg_reading, avg_writing
```

Identify Top 5 Students

```python
# Add a new column for total score
df['total_score'] = df['math score'] + df['reading score'] + df['writing score']
```

```python
# Sort and pick top 5
top5 = df.sort_values(by='total_score', ascending=False).head(5)
top5
```

Visualize Performance

```python
# Bar chart — Average marks
plt.figure(figsize=(6,4))
plt.bar(['Math','Reading','Writing'], [avg_math, avg_reading, avg_writing])
plt.title("Average Subject Scores")
plt.ylabel("Marks")
plt.show()
```

```python
# Histogram — Score distribution
sns.histplot(df['math score'], bins=20, kde=True)
plt.title("Math Score Distribution")
plt.show()
```

Insights

- Reading and Writing have higher average scores than Math.
- Math score distribution shows more low-performing students.
- Top 5 students have total scores above ~250.
- Overall performance shows students do well in language subjects.
