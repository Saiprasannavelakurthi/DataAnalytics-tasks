Category-wise Sales (Bar Chart)

```python
import matplotlib.pyplot as plt
import seaborn as sns

category_sales = df.groupby('PRODUCTLINE')['SALES'].sum()

plt.figure(figsize=(10, 6))
sns.barplot(x=category_sales.index, y=category_sales.values)
plt.title("Category-wise Sales")
plt.xlabel("Product Category")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```

Region-wise Sales (Pie Chart)

```python
region_sales = df.groupby('COUNTRY')['SALES'].sum()

plt.figure(figsize=(8,8))
plt.pie(region_sales.values, labels=region_sales.index, autopct='%1.1f%%')
plt.title("Region-wise Sales")
plt.show()
```

Top 5 Products by Sales

```python
top_products = (
    df.groupby('PRODUCTLINE')['SALES']
    .sum()
    .sort_values(ascending=False)
    .head(5)
)

top_products
```

Top 5 Regions by Sales

```python
top_regions = (
    df.groupby('COUNTRY')['SALES']
    .sum()
    .sort_values(ascending=False)
    .head(5)
)

top_regions
```

Visualize Top 5 Products

```python
plt.figure(figsize=(10,5))
sns.barplot(x=top_products.index, y=top_products.values)
plt.title("Top 5 Products by Sales")
plt.xticks(rotation=45)
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.show()
```

Visualize Top 5 Regions

```python
plt.figure(figsize=(10,5))
sns.barplot(x=top_regions.index, y=top_regions.values)
plt.title("Top 5 Regions by Sales")
plt.xticks(rotation=45)
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.show()
```
